```python
import numpy as np
import matplotlib.pyplot as plt
import math
from statistics import mean
import random
```

```python
m = 4.65e-26
temp = 298
boltz_k = 1.38 * 10**-23
sd = math.sqrt((temp*boltz_k)/m)

vx = np.random.normal(0, sd, 100000)
vy = np.random.normal(0, sd, 100000)
vz = np.random.normal(0, sd, 100000)

v = []
v2 = []
for i in range(len(vx)):
    v.append(math.sqrt(((vx[i])**2) + ((vy[i])**2) + ((vz[i])**2)))
    # v2.append(math.sqrt(((vx[i])**2) + ((vy[i])**2)))

print(mean(v))
plt.hist(v, rwidth=1, bins = 100)
plt.ylabel('time (s)')
plt.xlabel('total velocity (m/s)')
plt.title('Normal distribution v vs. t')
plt.show()
```
